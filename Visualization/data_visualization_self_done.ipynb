```python
import pandas as pd
import seaborn as sns
%matplotlib inline 
%config InlineBackend.figure_format = 'png'
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./Visualization_2.csv")

df.head(5)
```

```python
print("Missing Value Summary\n{}".format("-"*35))
print(df.isnull().sum(axis = 0))
```

## deal with NaN entries

```python
print(df.shape)

df = df.dropna(axis= 0)

print(df.shape)
```

```python
## make a pair plot
columns = ['age', 'num_streams']

axes = sns.pairplot(df,vars=columns,hue="is_subscriber",palette="husl")
```

```python
df.columns
```

```python
g =  sns.FacetGrid(data = df, row = 'country_name', col = 'subscriber_type', hue = "is_subscriber")
g.map(plt.scatter, 'num_streams', 'age').add_legend()
```

```python
g =  sns.FacetGrid(data = df, row = 'country_name', col = 'subscriber_type', hue = "is_subscriber")
g.map(plt.hist, 'num_streams').add_legend()
```

```python
g =  sns.FacetGrid(data = df, row = 'country_name', col = 'subscriber_type', hue = "is_subscriber")
g.map(plt.hist, 'age').add_legend()
```
